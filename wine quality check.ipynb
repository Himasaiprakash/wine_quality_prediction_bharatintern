```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("WineQT.csv")
```

```python
data
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split

x = data.drop(['quality'], axis=1)
y = data['quality']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
train_data = x_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(15,8))
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
```

```python
train_data['chlorides'] = np.log(train_data['chlorides']+1)
train_data['residual sugar'] = np.log(train_data['residual sugar']+1)
train_data['sulphates'] = np.log(train_data['sulphates']+1)
train_data['total sulfur dioxide'] = np.log(train_data['total sulfur dioxide']+1)
```

```python
train_data.hist(figsize=(15,8))
```
